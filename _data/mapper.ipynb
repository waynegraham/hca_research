# CLIR

```python
import pandas as pd
df = pd.read_csv('clir_members-geocoded.csv')
df.head()
```

```python
import geopandas as gpd

df["geometry"] = gpd.points_from_xy(df.longitude, df.latitude)
df.head()
```

```python
from shapely.geometry import Point
points = gpd.GeoSeries(df["geometry"][0]) # todo: fix this
centroid = points.centroid
points.centroid
```

```python
import folium

m = folium.Map(
    location=[points.centroid.y, points.centroid.x],
    zoom_start=3,
    tiles="Stamen Toner"
)

ala = folium.FeatureGroup(name="ALA Accredited Institutions")
for idx, row in df.iterrows():
    # folium.Marker(location=[row["latitude"], row["longitude"]], tooltip=row["name"]).add_to(m)
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]], 
        tooltip=row["institution"], 
        radius=10,
        color="#9C1944",
        fill=True,
        fill_color="#9C1944"
        ).add_to(m)

folium.LayerControl().add_to(m)
m.save('clir.html')
m
```

# Combined

Load the different datasets as a dataframe

```python
clir = pd.read_csv('clir_members-geocoded.csv')
hbcu = pd.read_csv('hbcu_members-geocoded.csv')
cenl = pd.read_csv('cenl_members-geocoded.csv')
rfn = pd.read_csv('rfn_members-geocoded.csv')
aau = pd.read_csv('aau_members-geocoded.csv')
```

Concatinate the dataframes to calculate a centroid

```python
from shapely.geometry import Point

# https://pandas.pydata.org/docs/user_guide/merging.html
frames = [clir, hbcu, cenl, rfn, aau]
result = pd.concat(frames)

result["geometry"] = gpd.points_from_xy(result.longitude, result.latitude)
points = gpd.GeoSeries(result["geometry"][0]) # todo: fix this


print(points)
```
